```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import nltk
from nltk.corpus import stopwords
nltk.download('all')
```

```python
# load data
emma = nltk.corpus.gutenberg.words('austen-sense.txt')

# remove non word and stopwords
emma = [w for w in emma if w.isalpha()]
emma = [w for w in emma if w not in stopwords.words('english')]
```

# How many total unique words are in the corpus

```python
# get unique word list
unique = set(emma)
print(len(emma),len(unique))
```

# Taking the most common words, how many unique words represent half of the total words in the corpus?

```python
# create frequency distribuction and get top 200 most frequent words
freq = nltk.FreqDist(emma)
emma_top200 = pd.DataFrame(freq.most_common(200),columns=['word','count'])
emma_top200['frequency'] = emma_top200['count']/float(len(emma))

emma_top200['toal frequency'] = emma_top200.cumsum()['frequency']
emma_top200.head(10)
```

```python
emma_top200[emma_top200['toal frequency']>=0.5].head(1)
```

```python
import matplotlib.pyplot as plot
plot.rcParams['figure.figsize'] = (21, 14)
freq.plot(65, cumulative=True)
```

Both graph and the data frame show that about 66 unique words account for half of the corpus
