```python
import nltk
import numpy as np
import pandas as pd
import random
%matplotlib inline
```

```python
data = 'https://raw.githubusercontent.com/andrewlulyj/Data-620/master/spambase.csv'
dataset= pd.read_csv(data)
```

```python
dataset.describe()
```

# Split Data Set 

```python
num = list(range(0,len(dataset)))
random.shuffle(num)
Train = num[0:3220]
Valid = num[3220:3910]
Test = num[3910:]
Train_set = dataset.loc[Train]
Valid_set = dataset.loc[Valid]
Test_set = dataset.loc[Test]
```

# Build Model


```python
# use only credit keyword only
def spam_features(line):
    return {'credit': line[19]}
```

```python
featureset_train = [(spam_features(row), row[-1]) for index, row in Train_set.iterrows()]
featureset_vaild = [(spam_features(row), row[-1]) for index, row in Valid_set.iterrows()]
featureset_test = [(spam_features(row), row[-1]) for index, row in Test_set.iterrows()]
```

```python
classifier = nltk.NaiveBayesClassifier.train(featureset_train)
print (nltk.classify.accuracy(classifier, featureset_vaild))
```

The accurancy against vaildation set is not that good

```python
# use credit , money and receive as keywords
def spam_features2(line):
    features = {}
    features["credit"] = line[19]
    features["money"] = line[23]
    features["receive"] = line[10]
    return features
```

```python
featureset2_train = [(spam_features2(row), row[-1]) for index, row in Train_set.iterrows()]
featureset2_vaild = [(spam_features2(row), row[-1]) for index, row in Valid_set.iterrows()]
featureset2_test = [(spam_features2(row), row[-1]) for index, row in Test_set.iterrows()]
```

```python
classifier2 = nltk.NaiveBayesClassifier.train(featureset2_train)
print (nltk.classify.accuracy(classifier2, featureset2_vaild))
```

after adding more features, the accurany has improved 

```python
# use credit money and will as keyword
def spam_features3(line):
    features = {}
    features["credit"] = line[19]
    features["money"] = line[23]
    features["will"] = line[11]
    return features
```

```python
featureset3_train = [(spam_features3(row), row[-1]) for index, row in Train_set.iterrows()]
featureset3_vaild = [(spam_features3(row), row[-1]) for index, row in Valid_set.iterrows()]
featureset3_test = [(spam_features3(row), row[-1]) for index, row in Test_set.iterrows()]
```

```python
classifier3 = nltk.NaiveBayesClassifier.train(featureset3_train)
print (nltk.classify.accuracy(classifier3, featureset3_vaild))
```

After replace receive with will the accurancy declined

```python
# use credit monry will and meeting as keywords
def spam_features4(line):
    features = {}
    features["credit"] = line[19]
    features["money"] = line[23]
    features["receive"] = line[10]
    features["meeting"] = line[41]
    return features
```

```python
featureset4_train = [(spam_features4(row), row[-1]) for index, row in Train_set.iterrows()]
featureset4_vaild = [(spam_features4(row), row[-1]) for index, row in Valid_set.iterrows()]
featureset4_test = [(spam_features4(row), row[-1]) for index, row in Test_set.iterrows()]
```

```python
classifier4 = nltk.NaiveBayesClassifier.train(featureset4_train)
print (nltk.classify.accuracy(classifier4, featureset4_vaild))
```

since receive seems to be a better feature than will, we decide to add it back and we add meeting into the feature set as well but 
the accurancy is still a little bit lower than model 2. so we decide to use model 2
